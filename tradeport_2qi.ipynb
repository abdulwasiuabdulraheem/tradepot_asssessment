```python
# Import neccesary librabries
from mlxtend.frequent_patterns import apriori, association_rules
import pandas as pd
import numpy as np

#load data
df = pd.read_excel("mkbasket.xlsx")
df.head(5)
```

```python
# Generate frequent itemsets
frequent_itemsets = apriori(df.drop(columns='Customer Name'), min_support=0.1, use_colnames=True)

# Generate association rules
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1)

# Sort rules by lift
rules = rules.sort_values(by='lift', ascending=False)

# Print top rules
rules.head()
```

MistaChoco and Sistachoco Promo Pack 15g x 50 and One Free Hanger of MistaChoco are the most likely variants to be bought together.
